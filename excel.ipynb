```python
import os
import pandas as pd

# Excel dosyalarının bulunduğu dizini belirtin
dizin = os.getcwd()

# Tüm Excel dosyalarını birleştirmek için boş bir DataFrame oluşturun
birlesik_df = pd.DataFrame()

# Dizindeki her Excel dosyasını bir DataFrame'e okuyun ve birleştirin
for dosya_adı in os.listdir(dizin):
    if dosya_adı.endswith('.xlsx'):  # Sadece Excel dosyalarını seçin (xlsx formatı için)
        dosya_yolu = os.path.join(dizin, dosya_adı)
        df = pd.read_excel(dosya_yolu)
        birlesik_df = pd.concat([birlesik_df, df], ignore_index=True)

# Tüm Excel verilerini içeren bir DataFrame elde edersiniz
print(birlesik_df)
```
